```python
Y=df.iloc[:,-1].values
X=df.iloc[:,:-1].values
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_new = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X_new,Y,test_size=0.3,random_state=1)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train,Y_train)
```

```python
svc.get_params()
```

```python
Y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_pred))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracy = []
for i in range(1,26):
    svc = SVC(C=i,kernel='rbf',random_state=1)
    score= cross_val_score(svc,X_train,Y_train,cv=5)
    accuracy.append(score.mean())
```

```python
plt.plot(range(1,26),accuracy)
plt.show()
```

```python
accuracy_1=[]
for i in range(1,26):
    svc = SVC(C=i,kernel='rbf',random_state=1)
    score = cross_val_score(svc,X_test,Y_test,cv=5)
    accuracy_1.append(score.mean())
```

```python
plt.plot(range(1,26),accuracy_1)
plt.show()
```

```python
accuracy_2 = []
for i in list(np.arange(0.1,5,0.1)):
    svc = SVC(C=i,kernel='rbf',random_state=1)
    score = cross_val_score(svc,X_train,Y_train,cv=5)
    accuracy_2.append(score.mean())
```

```python
plt.plot(list(np.arange(0.1,5,0.1)),accuracy_2)
plt.show()
```

```python
accuracy_3 = []
for i in list(np.arange(0.1,5,0.1)):
    svc = SVC(C=i,kernel='rbf',random_state=1)
    score = cross_val_score(svc,X_test,Y_test,cv=5)
    accuracy_3.append(score.mean())
```

```python
plt.plot(list(np.arange(0.1,5,0.1)),accuracy_3)
plt.show()
```

```python
gama = [0.001,0.01,0.1,1,10,100]
g_score = []
for i in gama:
    svc= SVC(C=1,kernel='rbf',random_state=1,gamma=i)
    score = cross_val_score(svc,X_train,Y_train,cv=5)
    g_score.append(score.mean())
```

```python
plt.plot(gama,g_score)
plt.grid()
plt.show()
```
